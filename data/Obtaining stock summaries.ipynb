```python
import csv
from bs4 import BeautifulSoup
import requests
from requests import get
import re
import json
```

```python
base_url = 'https://finance.yahoo.com/quote/{}/balance-sheet?p={}'

with open('financials.csv', 'w', newline='') as financials_file:
    with open('stocks.csv') as csvfile:
        next(csvfile, None)
        spamreader = csv.reader(csvfile, delimiter=',')
        spamwriter = csv.writer(financials_file, delimiter=',')
        spamwriter.writerow(['symbol', 'assets', 'shares'])
        
        for i, row in enumerate(spamreader):
            try:
                symbol = row[1]
                response = get(base_url.format(symbol, symbol))
                html_soup = BeautifulSoup(response.text, 'html.parser')
                print(i, symbol)

                m = re.findall('\"totalCurrentAssets.*?\}', response.text)
                js = json.loads('{' + m[-4] + '}')
                totalCurrentAssets = js['totalCurrentAssets']['raw']       
                print(totalCurrentAssets)

                m = re.findall('\"annualShareIssued.*?\}}]', response.text)
                js = json.loads('{' + m[0] + '}')
                shares = js['annualShareIssued'][-1]['reportedValue']['raw']
                print(shares)

                spamwriter.writerow([symbol, totalCurrentAssets, shares])
            except:
                pass
```
