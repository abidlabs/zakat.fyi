```python
import nltk
nltk.download('brown')
nltk.download('wordnet')
from nltk import FreqDist
from nltk.corpus import brown
frequency_list = FreqDist(i.lower() for i in brown.words())
```

```python
words, counts = zip(*frequency_list.items())
```

```python
sorted_words = [x for y, x in sorted(zip(counts, words))]
sorted_counts = [y for y, x in sorted(zip(counts, words))]

count_min = 150
count_max = 1000
len_word_min = 4
len_word_max = 10
```

```python
middle_words = [sorted_words[i] for i, count in enumerate(sorted_counts) if 
                (count_max > count > count_min) and (len_word_max > len(sorted_words[i]) > len_word_min)]
```

```python
from nltk.corpus import wordnet 
```

```python
all_key_words = []
all_taboo_words = []

for k, key_word in enumerate(middle_words):
    all_key_words.append(key_word)
    
    taboo_words = [] 
    for syn in wordnet.synsets(key_word): 
        taboo_words.extend([l.name() for l in syn.lemmas()])  # synonyms
        taboo_words.extend([l.antonyms()[0].name() for l in syn.lemmas() if l.antonyms()])  # antonyms 
        taboo_words.extend([l.name() for l in syn.hypernyms()])  # hypernyms 
        taboo_words.extend([l.name() for l in syn.hyponyms()])  # hyponyms 
    
    all_taboo_words.append(taboo_words)
```

Remove silly words and sort by frequency

```python
def clean_word(word):
    return word.split('.')[0]

def pick_top_words(key_word, word_list, num_words=5):
    diff_words = [w for w in word_list if (not(w==key_word) and not('_' in w))]
    return sorted(diff_words, key=lambda x: frequency_list.get(x, 0))[-num_words:]

def pick_bottom_words(key_word, word_list, num_words=5):
    diff_words = [w for w in word_list if (not(w==key_word) and not('_' in w))]
    return sorted(diff_words, key=lambda x: frequency_list.get(x, 0))[:num_words]
```

```python
for i, key_word in enumerate(all_key_words):
    taboo_words = [clean_word(w) for w in all_taboo_words[i]]
    top_taboo_words = pick_top_words(key_word, taboo_words)
    print(key_word, top_taboo_words)
```

# Other Approaches

```python
all_key_words = []
all_taboo_words = []

for k, key_word in enumerate(middle_words):
    all_key_words.append(key_word)
    
    taboo_words = [] 
    synonyms = []
    antonyms = []
    hypernyms = []
    hyponyms = []
    for syn in wordnet.synsets(key_word):           
        antonyms.extend([clean_word(l.antonyms()[0].name()) for l in syn.lemmas() if l.antonyms()])
        hypernyms.extend([clean_word(l.name()) for l in syn.hypernyms()])
        hyponyms.extend([clean_word(l.name()) for l in syn.hyponyms()])
        synonyms.extend([clean_word(l.name()) for l in syn.lemmas()])

    top_antonyms = pick_top_words(key_word, antonyms, 1)
    taboo_words.extend(top_antonyms)

    top_hypernyms = pick_top_words(key_word, hypernyms, 1)
    taboo_words.extend(top_hypernyms)

    top_hyponyms = pick_top_words(key_word, hyponyms, 1)
    taboo_words.extend(top_hyponyms)

    top_synonyms = pick_top_words(key_word, synonyms, 5-len(taboo_words))
    taboo_words.extend(top_synonyms)
                            
    all_taboo_words.append(taboo_words)
```

```python
with open('out_data/data4.csv', 'w', newline='') as output_csvfile:
    writer = csv.writer(output_csvfile) 
    for i, key_word in enumerate(all_key_words):
        if len(all_taboo_words[i]) == 5:
            print(key_word, all_taboo_words[i]) 
            writer.writerow([key_word, ','.join(all_taboo_words[i])])
```

# DataMuse Systematic

```python
import requests
import csv

with open('out_data/data4.csv', 'w', newline='') as output_csvfile:
    writer = csv.writer(output_csvfile) 
    for i, key_word in enumerate(all_key_words):
        top_taboo_words = []
        
        r_jjb = requests.get('https://api.datamuse.com/words?rel_jjb=' + key_word)
        jjb_taboo_words = [e['word'] for e in r_jjb.json()][:10]

        r_jja = requests.get('https://api.datamuse.com/words?rel_jja=' + key_word)
        jja_taboo_words = [e['word'] for e in r_jja.json()][:10]
        
        if len(jja_taboo_words) > len(jjb_taboo_words):
            taboo_words = jja_taboo_words
        else:
            taboo_words = jjb_taboo_words
        
        print(key_word)
        print(taboo_words)
        top_taboo_words = pick_bottom_words(key_word, taboo_words, 3)
        
        r = requests.get('https://api.datamuse.com/words?rel_trg=' + key_word)
        taboo_words = [e['word'] for e in r.json()]
        top_taboo_words.extend(pick_top_words(key_word, taboo_words, 2))
        
        print(top_taboo_words)
        
        if len(top_taboo_words) == 5:
            print('WRITTEN:', key_word, top_taboo_words) 
            writer.writerow([key_word, ','.join(top_taboo_words)])
```
